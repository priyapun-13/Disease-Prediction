```python
x=df.drop('Medical Condition',axis=1)
y=df['Medical Condition']
```

```python
model = LogisticRegression()
xtrain, xtest, ytrain, ytest = train_test_split(x, y , test_size=0.2 , random_state=123)
```

```python
print(xtrain.isnull().sum())
```

```python
xtrain = xtrain.fillna(xtrain.mean())
xtest = xtest.fillna(xtrain.mean()) 
```

```python
xtrain = xtrain.fillna(xtrain.median())
xtest = xtest.fillna(xtrain.median())
```

```python
for col in xtrain.select_dtypes(include=['object']).columns:
    xtrain[col] = xtrain[col].fillna(xtrain[col].mode()[0])
    xtest[col] = xtest[col].fillna(xtrain[col].mode()[0])
```

```python
xtrain = xtrain.dropna()
ytrain = ytrain[xtrain.index] 
```

```python
model.fit(xtrain, ytrain)
```

```python
y_pred=model.predict(xtest)
```

```python
print(y_pred)
```

```python
acc=accuracy_score(ytest,y_pred)
print (acc)
```

```python
print(confusion_matrix(ytest,y_pred))
```

```python
print(classification_report(ytest,y_pred))
```
